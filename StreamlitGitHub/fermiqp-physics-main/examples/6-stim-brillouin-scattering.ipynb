```python
import numpy as np
import matplotlib.pyplot as plt


import fermiqp_style as fqpstyle
fqpstyle.set_theme()
```

```python
def Pcr(Aeff,gb,L,alpha):
    Leff = (1-np.exp(-alpha*L))/alpha
    return 21*Aeff/(gb*Leff)
```

```python
loss_coeff_db = 2.5/1000 #db per m
loss_coeff_lin = loss_coeff_db/4.343
gain_br = 1e-11#brioullin gain in m/W
mfd = 12.4e-6/1.4 # mode field diameter in um (with fudge factor)

L = np.linspace(1,10,30)
plt.plot(L,Pcr(np.pi*mfd**2/4,gain_br,L,loss_coeff_lin),label='LMA-PM-15 (MFD=12.4e-6)')

mfd = 8.4e-6/1.4
plt.plot(L,Pcr(np.pi*mfd**2/4,gain_br,L,loss_coeff_lin),label='LMA-PM-10 (MFD=8.4e-6)')

plt.ylim(0,135)
plt.title('SBS power limit for PCF fibers @532nm')
plt.ylabel('SBS threshold (W)')
plt.xlabel('fiber length (m)')
plt.legend()
```
