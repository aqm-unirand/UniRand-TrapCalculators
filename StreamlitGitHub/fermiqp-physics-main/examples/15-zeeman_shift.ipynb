```python
from fermiqp.imports import *
from fermiqp.lattice import *
from fermiqp.experiments import *

import fermiqp_style as fqpstyle

from fermiqp_style.style import SPINDOWN, SPINUP

fqpstyle.set_theme()
from fermiqp.loaddata import load_zeeman
```

## RF transitions

```python
B, E = load_zeeman('2S')

y = E[:,1] - E[:,0]

plt.plot(B, y/1e6 - 80, '-', label='1-2')

y = E[:,2] - E[:,1]

plt.plot(B, y/1e6 - 80, '-', label='2-3')

plt.ylim(-10, 10)

plt.grid()

plt.xlabel('B (G)')
plt.ylabel('f / MHz - 80')

plt.legend()
```

## MW transitions

```python
B, E = load_zeeman('2S')

y = E[:,-1] - E[:,0]

plt.plot(B, y/1e9, '-', label='1-6')

y = E[:,-2] - E[:,1]

plt.plot(B, y/1e9, '-', label='2-5')

y = E[:,-3] - E[:,2]

plt.plot(B, y/1e9, '-', label='3-4')


plt.ylim(0, 3.5)

plt.grid()

plt.xlabel('B (G)')

plt.xlim(0, 1200)

plt.legend()

plt.ylabel('f [GHz]')
```

## Closed transitions

```python
Bg, Eg = load_zeeman('2S')
Be, Ee = load_zeeman('2P')
```

```python
d2_cool = Ee[0,:].max() - Eg[0,:].max()
d2_rep = Ee[0,:].max() - Eg[0,:].min()
d1_cool = Ee[0,:].min() - Eg[0,:].max()
d1_rep = Ee[0,:].min() - Eg[0,:].min()
ref = d2_cool
```

```python
y = Ee[:,6] - Eg[:,0] - ref

plt.plot(y/1e6, label=r'$|1\rangle - | m_J=-3/2,m_I =1\rangle$')

y = Ee[:,7] - Eg[:,1] - ref

plt.plot(y/1e6, label=r'$|2\rangle - | m_J=-3/2,m_I =0\rangle$')

y = Ee[:,8] - Eg[:,2] - ref

plt.plot(y/1e6, label=r'$|3\rangle - | m_J=-3/2,m_I =-1\rangle$')

plt.legend()

plt.grid()

plt.xlim(0, 1200)
plt.ylim(-1500, 228)

plt.ylabel('f - D2 cooler res [MHz]')

# plt.title('')
```

```python
y = Ee[:,-3] - Eg[:,-3] - ref

plt.plot(y/1e6, label=r'$|4\rangle - | m_J=3/2,m_I =1\rangle$')

y = Ee[:,-2] - Eg[:,-2] - ref

plt.plot(y/1e6, label=r'$|5\rangle - | m_J=3/2,m_I =0\rangle$')

y = Ee[:,-1] - Eg[:,-1] - ref

plt.plot(y/1e6, label=r'$|6\rangle - | m_J=3/2,m_I =-1\rangle$')

plt.legend()

plt.grid()

plt.xlim(0, 1200)
plt.ylim(0, 1600)

plt.ylabel('f - D2 cooler res [MHz]')

# plt.title('')
```
