```python
folder = 'Z:\\lab\\planning\\objective\\VisibleObjective_WFE_Different_Wavelengths'
files=[
    'Visible_Objective_Continous_Wavelengths_WFE.txt',
    'Visible_Objective_Continous_Wavelengths_Strehl.txt',
    'Visible_Objective_Continous_Wavelengths_SpotSize.txt'
]


fields = [0,29,40,50,75,100]


fig,axs=plt.subplots(1,3,figsize=(15,4))

labels = ['RMS Wavefront Error (waves)',
          'Strehl Ratio',
          'Spot Size ($\mu m$)']

for ax,file, label in zip(axs,files,labels):
    file_to_open = os.path.join(folder,file)
    data = pd.read_csv(file_to_open, delimiter='\t',encoding='utf-16', header = 7,decimal=",")
    
    for i in range(len(data.columns)-1):
        ax.plot(data[data.columns[0]].values,
                data[data.columns[i+1]].values,
                marker='',label=fields[i])
    ax.set_ylabel(label)
    ax.set_xlabel('Wavelenth ($\mu m$)')
    ax.legend(title='Field ($\mu m$)')

axs[0].axhline(1/14,marker='',ls='--',c='grey')
axs[1].axhline(0.8,marker='',ls='--',c='grey')
```

```python
folder = 'Z:\\lab\\planning\\objective\\VisibleObjective_WFE_Different_Wavelengths'
files=[
    'Visible_Objective_Continous_Wavelengths_WFE.txt',
    'Visible_Objective_Continous_Wavelengths_Strehl.txt',
    'Visible_Objective_Continous_Wavelengths_SpotSize.txt'
]


fields = [0,29,40,50,75,100]


fig,axs=plt.subplots(3,1,sharex=True,figsize=(10,10))

labels = ['RMS Wavefront Error (waves)',
          'Strehl Ratio',
          'Spot Size ($\mathrm{\mu m}$)']

for ax,file, label in zip(axs,files,labels):
    file_to_open = os.path.join(folder,file)
    data = pd.read_csv(file_to_open, delimiter='\t',encoding='utf-16', header = 7,decimal=",")
    
    for i in range(len(data.columns)-1):
        ax.plot(data[data.columns[0]].values,
                data[data.columns[i+1]].values,
                marker='',label=fields[i])
    ax.set_ylabel(label)
    ax.grid()
    
axs[1].legend(title='Field ($\mathrm{\mu m})$',loc='lower right')#bbox_to_anchor=(1.2,0.5))
axs[2].set_xlabel('Wavelength ($\mathrm{\mu m}$)')
axs[0].axhline(1/14,marker='',ls='--',c='grey')

axs[1].axhline(0.8,marker='',ls='--',c='grey')
axs[2].plot(data[data.columns[0]].values,2*data[data.columns[0]].values/(2*0.65)*1.22,marker='',ls='--',c='grey')
plt.subplots_adjust(wspace=0, hspace=0)

plt.suptitle('Top Objective Performance at Different Wavelengths')
plt.tight_layout()
```
