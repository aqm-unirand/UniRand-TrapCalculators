## Pinning

```python
pinn = PinningLattice(waist=60e-6)

x = np.linspace(-1, 1, 100) * 1e-6

X, Y = np.meshgrid(x, x)

V = pinn.potential(X, Y)

plt.pcolormesh(X, Y, V)
plt.gca().set_aspect('equal')
```

## t' configuration with a waveplate

```python
theta = np.pi/2 + 0.1

pinn = PinningLattice(waist=60e-6, amplitudes = [1, np.cos(theta), np.cos(theta), 1, np.sin(theta), np.sin(theta)])
pinn_geo = GeometryBeam(waist=60e-6, amplitudes=[0, 0])

x = np.linspace(-1., 1., 100) * 1e-6 * 1.3

X, Y = np.meshgrid(x, x)

Es = pinn.scalar_field(X, Y, pol='s')
Ep = pinn.scalar_field(X, Y, pol='p')

V = pinn.potential(X, Y)
V += pinn_geo.potential(X, Y)

plt.pcolormesh(X, Y, V, shading='auto')
plt.gca().set_aspect('equal')
```
